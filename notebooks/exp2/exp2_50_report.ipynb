## Abstract

* We hypothesize that a variant of BOCD might be able to segment EEG data into neurotypical brain states related to epilepsy neurodynamics.
* To this end, we apply BOCD to raw EEG.
* As a proof of concept, we apply the algorithm to seizure EEG, on each channel separately.
* In this experiment the BOCD algorithm returned zero changepoints.
* Further tasks include varying the default BOCD parameters (LAMBDA, ALPHA, BETA, KAPPA, MU, THRESHOLD).

* LAMBDA = 100
* ALPHA = 0.1
* BETA = 1.
* KAPPA = 1.
* MU = 0.
