## Imports

```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

%matplotlib inline
```

## Define the initial value problem
Excitable System (Strogatz 4.5.3):

$\dot{\theta} = \mu + \sin {\theta}$, where $\mu \approx 1-\epsilon$

Or equivalently:

$\dot{y} = f(t, y) = a + \sin(y)$



```python
def f(t, y, a=0.95):
    ydot = a + np.sin(y)
    return np.r_[ydot]

fig, ax = plt.subplots(1,1, figsize=(12,4))

# we want to evaluate the system on 30 linearly
# spaced times between t=0 and t=10
t = np.linspace(0., 10., 50)

# we simulate the system for different values of a
for a in np.linspace(0.8, 0.98, 5):
    # the initial position is (pi/2)
    y0 = np.array([np.pi/2])
    # y0 = np.array([np.arcsin(-a)])  # defining y0 to be a fixed point
    # we simulate the system and evaluate $y$ on the given times
    sol = spi.solve_ivp(f, t, y0, args=(a,), dense_output=True)
    y = sol.sol(t)
    # we plot the particle's trajectory
    ax.plot(t, y.T, mew=1, ms=8,
            mec='w', label=f'{a:.2f}')
ax.set_title('y')
ax.legend()
```

## Define membrane potential
$V(t) = \cos \theta (t)$

```python
def v(y):
    return np.cos(y)

plt.plot(t, v(y.T))

```
