## Imports


```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt

%matplotlib inline
```

## Define the initial value problem
Excitable System (Strogatz):

$\dot{\theta} = \mu + \sin {\theta}$, where $\mu \approx 1-\epsilon$

Or equivalently:

$\dot{y} = f(t, y) = a + \sin(y)$

```python
a = 0.95
# the particle's state is (y)
# the initial position is (pi/2)
y0 = np.array([np.pi/2])

def f(t, y, a=0.95):
    ydot = a + np.sin(y)
    return np.r_[ydot]

fig, ax = plt.subplots(1,1, figsize=(12,4))

# we want to evaluate the system on 30 linearly
# spaced times between t=0 and t=3
t = np.linspace(0., 3., 30)

# we simulate the system for different values of a
for a in np.linspace(0.8, 1, 5):
    # we simulate the system and evaluate $y$ on the given times
    sol = spi.solve_ivp(f, t, y0, args=(a,), dense_output=True)
    y = sol.sol(t)
    # we plot the particle's trajectory
    ax.plot(t, y.T, mew=1, ms=8,
            mec='w', label=f'{a:.2f}')

ax.legend()
```
